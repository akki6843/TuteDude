**Task 1 - Basic Function: Price After Discount**

Write a function apply_discount(price, discount_percent) that:
1. Returns the price after discount.
2. If discount_percent is missing, apply a default discount of 5%.
3. Test the function with:
4. apply_discount(1000, 10)
5. apply_discount(500) # uses default discount

Extra (optional): Add a condition inside the function to ensure discount never exceeds 60%.

```python
def apply_discount(price, discount_percent=5):
    # Extra: ensure discount never exceeds 60%
    if discount_percent > 60:
        discount_percent = 60

    discounted_price = price - (price * discount_percent / 100)
    return discounted_price



print("Price after 10% discount on 1000:", apply_discount(1000, 10))
print("Price after default discount on 500:", apply_discount(500))  # uses default 5%
print("Price after (excessive) 80% discount on 1000:", apply_discount(1000, 80))  # capped at 60%
```

Explanations :  
- Implementation of Default Argument Done.
- Tested with given arguments.
- implemented 60% cap on discount using if statement.

**Task 2 - Recursive Function: Factorial Utility**  
Create a recursive function factorial(n) that:  

- Returns the factorial of n.  
- Handles two edge cases: n == 0 and n == 1.  
- Prints an error message if n is negative.  
  
Test with:  
factorial(5)  
factorial(0)  
factorial(-3)  

```python
def factorial(n):
    # Handle negative input
    if n < 0:
        print("Error: Factorial is not defined for negative numbers.")
        return None

    # Base cases
    if n == 0 or n == 1:
        return 1

    # Recursive case
    return n * factorial(n - 1)


# Testing the function
print("factorial(5):", factorial(5))
print("factorial(0):", factorial(0))
print("factorial(-3):", factorial(-3))
```

Explanation :   
- Recursion is a unique type of function, where they call themself for the solution.  
- Important thing to keep in find is the base or exit case for the recursion to excute perfectly.

**Task 3 - Lambda Function: GST Calculator**   
Create a lambda function gst that returns price after adding 18% GST.  
Example:  

`gst = lambda price: price + (0.18 * price)`  
`print(gst(100)) # should return 118 `  
  
Extra (optional): Create another lambda to compute final price after GST + discount together.  

```python
# Basic lambda: price after 18% GST
gst = lambda price: price + (0.18 * price)

print("Price after GST on 100:", gst(100))  # should return 118
print("Price after GST on 500:", gst(500))


# Extra: lambda to compute final price after discount + GST together
final_price = lambda price, discount_percent: (price - (price * discount_percent / 100)) + \
                                               (0.18 * (price - (price * discount_percent / 100)))

print("Final price (1000, 10% discount, then GST):", final_price(1000, 10))
```

Explanations :   
- lambda functions are special type of function, they are also kind of nameless function.  
- One key thing about lambda is its signature `lambda input: output`.  
- Any kind of quick process that can be done on output lambda function is good to be used.

**Task 5 - Using filter(): Filter Expensive Products**  
Given the list:  
`prices = [100, 250, 400, 1200, 50, 2000, 850]`  
Use filter() to:  

Create a list of prices greater than 500.  
Create another list of prices less than or equal to 500.  

Print both lists.  

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]

# Prices greater than 500
expensive_prices = list(filter(lambda price: price > 500, prices))

# Prices less than or equal to 500
affordable_prices = list(filter(lambda price: price <= 500, prices))

print("Prices greater than 500:", expensive_prices)
print("Prices less than or equal to 500:", affordable_prices)
```

Explanations :   
- filter() vs map(): filter() keeps only the elements where the lambda returns True, rather than transforming every element like map() does — so the output list can be shorter than the input, which is exactly what you want here.

**Task 6 - Combined Utility Function**  
Write a function process_prices(prices) that:  

Takes a list of prices.  
Uses map() + lambda to apply 10% discount to all prices.  
Uses filter() to keep only discounted prices that are above 300.  
Returns both lists:  

discounted_prices  
filtered_prices  



Test with:  
`process_prices([100, 500, 900, 50, 750])`

```python
def process_prices(prices):
    # Apply 10% discount to all prices using map() + lambda
    discounted_prices = list(map(lambda price: price - (price * 0.10), prices))

    # Keep only discounted prices above 300 using filter()
    filtered_prices = list(filter(lambda price: price > 300, discounted_prices))

    return discounted_prices, filtered_prices


# Testing the function
discounted, filtered = process_prices([100, 500, 900, 50, 750])
print("Discounted prices:", discounted)
print("Filtered prices (above 300):", filtered)
```

- Returning two lists:  
    Return discounted_prices, filtered_prices returns them as a tuple, so unpacking with discounted, `filtered = process_prices(...)` on the caller side is the clean way to receive both — this matches the "returns both lists" requirement without needing classes or dictionaries.    
- Filter operates on the discounted values, not the originals:  
    The filter step in filtered_prices runs on discounted_prices (the post-discount amounts), since the task says "keep only discounted prices that are above 300" — worth double-checking that's the intended reading, as opposed to filtering the original prices above 300 and then discounting those.

**Task 7 - Mini Problem: Menu Using Functions**  
Create the following three small functions:  

- add_price(prices_list, price) → adds price to the list.  
- get_average_price(prices_list) → returns average price.  
- get_max_price(prices_list) → returns the maximum price.  

Then create a simple menu:  

`1 → Add price  
2 → Show average price  
3 → Show highest price  
q → Quit`  

Use only loops + function calls (no OOP).

```python
def add_price(prices_list, price):
    prices_list.append(price)
    return prices_list


def get_average_price(prices_list):
    if len(prices_list) == 0:
        return 0
    return sum(prices_list) / len(prices_list)


def get_max_price(prices_list):
    if len(prices_list) == 0:
        return 0
    return max(prices_list)


prices_list = []

while True:
    print("\n--- Menu ---")
    print("1 -> Add price")
    print("2 -> Show average price")
    print("3 -> Show highest price")
    print("q -> Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        price_input = input("Enter price to add: ")
        if price_input.replace('.', '', 1).isdigit():
            add_price(prices_list, float(price_input))
            print("Price added. Current list:", prices_list)
        else:
            print("Invalid price. Please enter a valid number.")

    elif choice == "2":
        print("Average price:", get_average_price(prices_list))

    elif choice == "3":
        print("Highest price:", get_max_price(prices_list))

    elif choice == "q":
        print("Exiting menu. Goodbye!")
        break

    else:
        print("Invalid choice. Please select 1, 2, 3, or q.")
```

Explanations :   
- No try/except used for price validation — 
    Since the restrictions rule out exceptions, I validated the input with price_input.replace('.', '', 1).isdigit(), which checks the string looks like a valid non-negative number (allowing at most one decimal point) before converting it with float(). 
    This is a common pattern for validating numeric strings without exceptions, though it won't catch negative numbers (e.g., -5 fails isdigit() too, so it's actually rejected — that's fine here since negative prices don't make sense anyway).
- While True + break is the loop mechanism for the menu — 
    It keeps prompting until the user types q, satisfying "use only loops."
- Functions stay small and single-purpose, each handling exactly one job (add_price, get_average_price, get_max_price) and getting called from the menu loop, rather than embedding logic directly inside the if/elif chain.
