**Task 1: Write Sales Records to a File**  


Create a list of sales amounts:  

`sales = [1200, 450, 980, 1500, 3000]`  
Open a file named sales_data.txt in write mode.  
Write each sale on a new line in the file.  
Close the file, then reopen it and print its contents.  
Extra (optional): Write the data in comma-separated format instead of separate lines.  

```python
# Step 1: Create the list
sales = [1200, 450, 980, 1500, 3000]

# Step 2 & 3: Write sales to file
with open("sales_data.txt", "w") as file:
    for sale in sales:
        file.write(str(sale) + "\n")

# Step 4: Read and print contents
with open("sales_data.txt", "r") as file:
    print("Contents of sales_data.txt:")
    print(file.read())
```

```python
# Optional CSV format
sales = [1200, 450, 980, 1500, 3000]

with open("sales_data.csv", "w") as file:
    file.write(",".join(map(str, sales)))

with open("sales_data.csv", "r") as file:
    print(file.read())
```

Explanation :  
- Since data has to be written to a file everything is converted to string.  
- Onlt difference between Normal and CSV approach is the use of comma for creating a string.

**Task 2: Read File in Different Ways**  
  
Using the same sales_data.txt:  
  
Read the entire file using .read() and print it.  
Read the first line using .readline().  
Read all lines using .readlines() and convert them into a list of integers.  
Ensure proper formatting and cleanup of newline characters.  

```python
# 1. Read the Entire File :

with open("sales_data.txt", "r") as file:
    data = file.read()

print("Entire File:")
print(data)
```

```python
# 2. Reading only the first line of the file:

with open("sales_data.txt", "r") as file:
    first_line = file.readline()

print("First Line:")
print(first_line.strip())
```

```python
# 3. Reading all lines into a list AND THEN converting them to integers:

sales_list = []
with open("sales_data.txt", "r") as file:
    for line in file:
        # Split by comma and convert each piece to int
        numbers = [int(x) for x in line.strip().split(",")]
        sales_list.extend(numbers)

print("Sales List:")
print(sales_list)
```

**Task 3: Append New Sales**  

Append these new sales to the same file:  
`
5000
2500
1700
`  
After appending, reopen and print the entire updated file.  
Extra (optional): Print the total number of lines after appending.  

```python
# 1 : Append : 

new_sales = [5000, 2500, 1700]

with open("sales_data.txt", "a") as file:
    for sale in new_sales:
        file.write(str(sale) + "\n")
```

```python
# 2 : Read Updated File :

with open("sales_data.txt", "r") as file:
    print("Updated File:")
    print(file.read())
```

```python
# 3 : Cout total number of lines in the file :

with open("sales_data.txt", "r") as file:
    lines = file.readlines()

print("Total number of lines:", len(lines))
```

**Task 4: Generate Summary Report from File**  
  
Using only file reading operations:  

1. Read all sales values from sales_data.txt.  
2. Convert them into integers.  
3. Calculate and print:  
- Total Sales  
- Highest Sale  
- Lowest Sale  
- Average Sale  
4. Do not use any advanced libraries.  

```python
with open("sales_data.txt", "r") as file:
    sales = [int(line.strip()) for line in file.readlines()]


total_sales = sum(sales)
highest_sale = max(sales)
lowest_sale = min(sales)
average_sale = total_sales / len(sales)


print("Sales Summary Report")
print("--------------------")
print("Total Sales:", total_sales)
print("Highest Sale:", highest_sale)
print("Lowest Sale:", lowest_sale)
print("Average Sale:", average_sale)
```

**Task 5: Create Product Info File (User Input)**  
  
1. Ask the user for 3 product names and their prices.  
2. Write them into a new file products.txt in this format:  

    `ProductName | Price`  
3. Read the file and print each line with proper formatting.

```python
# Create products.txt
with open("products.txt", "w") as file:
    for i in range(3):
        product = input("Enter product name: ")
        price = input("Enter product price: ")
        file.write(product + " | " + price + "\n")

# Read and display the file
print("\nProduct List")
print("------------")

with open("products.txt", "r") as file:
    for line in file:
        print(line.strip())
```

**Task 7: Mini Project - Export Discounted Prices**  


Create a dictionary:

`prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}`
  
Ask the user for a discount percentage.
  
Write discounted prices into discount_report.txt using the format:  
  
`Product | Original Price | Discounted Price`

After writing, read the file and print it to the terminal.  
  
Extra (optional): Write a summary at the bottom of the file:  
`
Total Items: X
Average Discounted Price: Y
`

```python
# Product prices
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

# Get discount percentage
discount = float(input("Enter discount percentage: "))

total_discounted = 0

# Write discounted prices
with open("discount_report.txt", "w") as file:
    file.write("Product | Original Price | Discounted Price\n")

    for product, price in prices.items():
        discounted_price = price - (price * discount / 100)
        total_discounted += discounted_price

        file.write(f"{product} | {price} | {discounted_price:.2f}\n")

    # Optional Summary
    file.write("\n")
    file.write(f"Total Items: {len(prices)}\n")
    file.write(f"Average Discounted Price: {total_discounted / len(prices):.2f}\n")

# Read and display the report
print("\nDiscount Report")
print("----------------")

with open("discount_report.txt", "r") as file:
    print(file.read())
```
