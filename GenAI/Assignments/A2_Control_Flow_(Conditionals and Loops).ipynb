## **Task 1: Discount Rules (if / elif / else)**

Write a program that reads an integer order_amount from the user using input().  
Apply the following discount rules and print the final amount:  
- order_amount >= 2000 → 15% discount  
- 1500 <= order_amount < 2000 → 10% discount  
- 1000 <= order_amount < 1500 → 7% discount  
- otherwise → 0% discount  
Ensure you convert input to a number and handle the case when the input is non-numeric by displaying an error message and exiting the program.  
Extra (optional): Add tax (fixed 5%) after discount and print subtotal, tax, and final total (still within conditionals and basic arithmetic).  

```python
'''
Collecting the order amount from the user and handling invalid input.
'''
try :
    order_amount = int(input("Enter the order amount: "))
except ValueError:
    print("Invalid input. Please enter a valid integer.")
    # exit()  #--------------------> To stop the further execution of the program in case of invalid input, uncomment this line. But this will terminate the kernal and you have to restart the kernal to run the program again. So, I have commented it out for now.
    order_amount = 0

if order_amount < 0:
    print("Order amount cannot be negative. Please enter a valid amount.")
    order_amount = 0

print("Order amount:", order_amount)
```

```python
'''
Implementing the discount logic based on the order amount.
'''
if order_amount >= 2000:
    discount_rate = 0.15
elif order_amount >= 1500:  # Since we already checked >=2000, this implies <2000
    discount_rate = 0.10
elif order_amount >= 1000:  # Implies >=1000 and <1500
    discount_rate = 0.07
else:  # Otherwise (<1000)
    discount_rate = 0.00


discount_amount = order_amount * discount_rate
final_amount_after_discount = order_amount - discount_amount

print(f"Discount applied: {discount_rate*100}%")
print(f"Final amount after discount: {final_amount_after_discount:.2f}")
```

```python
'''
Implementing optional Tax logic.
'''
tax_rate = 0.05
tax_amount = final_amount_after_discount * tax_rate
total_with_tax = final_amount_after_discount + tax_amount

print("-" * 40)
print("With Tax Calculation (Extra):")
print(f"Subtotal (after discount): {final_amount_after_discount:.2f}")
print(f"Tax (5%): {tax_amount:.2f}")
print(f"Final Total: {total_with_tax:.2f}")
print("-" * 40)
```

**Explanation :**  
  1 : Collect order amount from the user, It has 2 checks:  
        - If its a valid integer input or not.    
        - If the order is possitive or not.  
    
  2 : Implementation of Discounts as per the requirement:  
        - Get the discount percent, multiply it with order, to get final discount value.  
        - Calculate final bill by substracting the discount. 
    
  3 : Finally, applying Tax to the discounted amount.   
        - apply 5% tax to final amount  
        - Add the tax to bill amount to get full and final bill.  
  
  **NOTE** : While excepting the input from the user requirments are to stop the code, since this is notebook implementation, I have commented the part but if in-case you want to experience it uncomment the `exit()` and it will work.

## **Task 2: Process Multiple Orders (for loop)**

1. Given a list of order amounts: `orders = [1200, 2500, 800, 1750, 3000]`, use a for loop to apply the discount rules from Task 1 to each order and print a summary table showing: `order_amount -> discount% -> final_amount`.

2. Also compute and print the total revenue after discounts.

Extra (optional): Print the number of orders that received a discount (`discount > 0`).

```python
order_amounts = [1200, 2500, 800, 1750, 3000]

orders_received_discount = 0

for order_amount in order_amounts:
    if order_amount >= 2000:
        discount_rate = 0.15
    elif order_amount >= 1500:  # Since we already checked >=2000, this implies <2000
        discount_rate = 0.10
    elif order_amount >= 1000:  # Implies >=1000 and <1500
        discount_rate = 0.07
    else:  # Otherwise (<1000)
        discount_rate = 0.00

    if discount_rate > 0:
        orders_received_discount += 1
    discount_amount = order_amount * discount_rate
    final_amount_after_discount = order_amount - discount_amount

    print(f"Order amount: {order_amount}")
    print(f"Discount applied: {discount_rate*100}%")
    print(f"Final amount after discount: {final_amount_after_discount:.2f}")
    print("-" * 40)

print('')
print(f"Total orders received discount: {orders_received_discount} out of {len(order_amounts)}") 
```

**Explanations :**  

- Loop through order amouts apply the discount and then get final order value.  
- Finally printing the number of orders that got discount.
- Personally I would have implemneted a function to get the discount percent, it would have made code cleaner and leaner. But Strict No function use was instructed.

## **Task 3: User Menu (while loop + break/continue)**

1. Implement a simple while-loop driven menu that keeps asking the user to choose an action until they press q to quit.
Menu options:
   - 1 — Add order amount to a running list (use input())
   - 2 — Show all orders and totals after applying discounts
   - q — Quit

2. Use continue to re-show the menu after invalid input and break to exit on q.

Constraints: Keep the loop logic simple and use only lists and loops (no functions required).

```python
orders = []   # running list of order amounts

while True:
    # ---- show the menu every iteration ----
    print("\n===== Order Menu =====")
    print("1 — Add order amount")
    print("2 — Show all orders and totals after discounts")
    print("q — Quit")
    print("======================")

    choice = input("Choose an option: ").strip().lower()

    # ---- add order amount ----
    if choice == '1':
        try:
            order_amount = float(input("Enter the order amount: "))
            if order_amount < 0:
                print("Error: Amount cannot be negative.")
                continue
            orders.append(order_amount)
            print(f"Order amount {order_amount} added.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")
            continue

    # ---- show all orders and totals after discounts ----
    elif choice == '2':
        if not orders:
            print("No orders to display.")
            continue
        
        else:
            print("\n--- ORDER SUMMARY ---")
            print(f"{'Order Amount':<15} {'Discount %':<12} {'Final Amount':<15}")
            print("-" * 42)

            for order_amount in orders:
                if order_amount >= 2000:
                    discount_rate = 0.15
                elif order_amount >= 1500:
                    discount_rate = 0.10
                elif order_amount >= 1000:
                    discount_rate = 0.07
                else:
                    discount_rate = 0.00

                discount_amount = order_amount * discount_rate
                final_amount_after_discount = order_amount - discount_amount
                print(f"{order_amount:<14.2f} {discount_rate*100:<11.0f} {final_amount_after_discount:<14.2f}")
                print("-" * 42)

        

    # ---- quit ----
    if choice == 'q':
        print("Goodbye!")
        break                       # exit the while loop on 'q'
```

Explanations :   
while loop implementation to mimic constant order summary.  
- By pressing `1` Keep accepting order amount.   
- If user wants to see all the orders press `2`.
- And finally `q` to exit the menu.

## **Task 4: Loop Control with Conditions (break & continue)**

1. Given a list of daily sales: `daily = [200, 150, 0, 400, 50, -1, 300]`, iterate through it with a `for` loop and:
   - If a day's value is `-1`, treat it as corrupted data and `break` the loop (stop processing).
   - If a day's value is `0`, treat it as a day with no sales and `continue` (skip adding to revenue).
   - For valid positive sales, add to `total_sales` and print the running total.

2. Print the final total after the loop completes (or stops due to corrupted data).

```python
daily = [200, 150, 0, 400, 50, -1, 300]

total_sales = 0

for amount in daily:
    if amount < 0:
        print(f"Invalid order amount: {amount}. Data Corrupted.")
        break  # Exit the loop on negative amount
    elif amount == 0:
        print("Order amount is zero. No Sale registered.")
        continue
    total_sales += amount
    print(f"Added {amount} to total sales. Current total: {total_sales}\n")

print('')
print(f"Total sales for the day: {total_sales}")
```

Explanation :   
- 2 condistions of corrupt data and zero sales implemented using if, elif statement.  
- then just adding running total and then displaying the total sales.
